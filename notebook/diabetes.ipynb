## Introduktion
### Om datasettet

```
https://www.kaggle.com/datasets/alexteboul/diabetes-health-indicators-dataset

```

Datasettet är baserat på ett dataset från Center of Disease Control and Prevention(CDC), den amerikanske motsvarigheten till Folkhälsomyndigheten. CDC sammanställer årligen data från det de kallar Behavioral Risk Factor Surveillance System (BRFSS). Det datasettet innehåller 454 kolumner, medan detta är städat och sammanställt med de parametrar som denna person såg mest relevanta för klassificering av diabetesrisk.
Datasettet innhehåller 21 kolumner varav en är en binär target-kolumn. 0 för ingen diabetes och 1 för pre-diabetes eller diabetes och datasettet är jämt balanserat mellan de båda klasserna. Det finns ingen information i datasettet om vilken typ av diabetes som avses. I och med att till exempel arv eller graviditet inte finns med som parametrar så har jag utgått från att det är diabetes typ 2 som avses. Risken att utveckla diabetes typ 2 hänger till stor del ihop med levnadsvanor, vilket många av de ingående kolumnerna kan tänkas vara kopplade till. 

### Problem
Kan jag träna en klassificeringsmodell som tillförlitligt kan identifiera personer som har en ökad risk för att utveckla diabetes typ 2.

### Mål
Målet med projektet är att träna en maskininlärningsmodell som tillförlitligt kan kategorisera personer som befinner sig i riskzon för att utveckla diabetes. Träning av modellen för att nå målsättningen kan innebära att personer som inte har någon egentlig risk feldiagnosticeras som riskpersoner. Modellen är dock endast tänkt som ett hjälpmedel för en snabb första bedömning innan vidare utredning, därför ser jag detta som en acceptabel kompromiss.

### Positiv korrelation 
**GenHlth** - Tycker du att du har en god allmän hälsa? 1 är Utmärkt och 5 är Dålig. Sämre egenbedömd hälsa har alltså en positiv korrelation med risk för diabetes, vilket jag finner logiskt då flera faktorer påverkar risken för diabetes och en generell bedömning kan fånga in flera av dessa faktorer.  
**HighBP** - Högt blodtryck? 0 för nej, 1 för ja. Högt blodtryck har en positiv korrelation med risk för diabetes. Högt blodtryck kan tyda på att där finns andra faktorer som kan öka risk för diabetes typ 2, till exempel fetma eller övervikt, dålig kosthållning mm.

**Age** - Högre ålder har en positiv korrelation med risken för att insjukna i diabetes. Diabetes tar lång tid att utveckla varför fler fall upptäcks i högre ålderskategorier.  
**DiffWalk** - Har du stora problem att gå eller gå i trappor? 0 för nej, 1 för ja. Problem att gå har en positiv korrelation med risk att insjukna i diabetes. Flera av de andra faktorerna kan påverkas om personen inte kan vara fysisk aktiv. Till exempel kan BMI, allmänhälsan och blodtrycket påverkas negativt.  
**PhysHlth** - Hur många av de senaste 30 dagarna har du varit skadad/sjuk? Skador och sjukdomar kan påverka och påverkas av faktorer som även ökar risk för diabetes.  
**HeartDiseaseorAttack** - Dokumenterade hjärt- och kärlsjukdomar har en positiv korrelation med risk för diabetes. Det är vanligare att få hjärtattack om man har diabetes.  

### Negativ korrelation

**Education** - Högre utbildningsnivå har negativ korrelation med risk för diabetes. Det kan bero på att personer med lägre utbildningsnivå har mindre kunskap om riskerna med ett osunt levnadssätt.  
**Income** - Högre inkomstnivå har negativ korrelation med risk för diabetes. Eftersom datan kommer från USA kan detta visa på skillnader mellan hur låg- respektive höginkomsttagare är sjukvårdsförsäkrade, där de med högre lön kanske till högre grad har råd att gå på kontroller.

## Förbereda data för modellträning

### Separera target från features samt skapa tränings- och testdata.

````python
X_filtered = df_filtered.drop(['Diabetes_binary'], axis=1)
y_filtered = df_filtered['Diabetes_binary']

X_train_filtered, X_test_filtered, y_train_filtered, y_test_filtered = train_test_split(X_filtered, y_filtered, test_size=0.2, random_state=42, stratify=y_filtered)
````

### Skala features för att minska risken för att någon feature får större vikt
Värdena i feature-kolumnerna har olika skalor, t ex 0-30 för PhysHlth och 1-5 för GenHlth. Genom att använda StandardScaler skalas alla kolumner om till en skala med medelvärdet 0 och standardavvikelsen 1.

````python
scaler = StandardScaler()

X_train_scaled_filtered = scaler.fit_transform(X_train_filtered)
X_test_scaled_filtered = scaler.fit_transform(X_test_filtered)
X_test_scaled_filtered  = pd.DataFrame(X_test_scaled_filtered, columns=X_filtered.columns)
X_train_scaled_filtered  = pd.DataFrame(X_train_scaled_filtered, columns=X_filtered.columns)
````

## Val av maskininlärningsmodeller

Med målet att skapa en klassificeringsmodell faller valet på att testa RandomForestClassifier, ExtraTreesClassifier samt GradientBoostingClassifier.


Teststrategin blir att se hur de tre modellerna presterar i sitt default-utförande och sedan försöka optimera recall-1 i den modell som presterat bäst. 



### Gradient Boosting Classifier

````python
gradient_boosting_classifier = GradientBoostingClassifier(random_state=42)
gradient_boosting_classifier.fit(X_train_scaled_filtered, y_train_filtered)

gradient_boosting_classifier_pred = gradient_boosting_classifier.predict(X_test_scaled_filtered)
````

````python
cm_gb = confusion_matrix(y_test_filtered, gradient_boosting_classifier_pred)
plt.figure(figsize=(6,5))
sns.heatmap(cm_gb, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix - Gradient Boosting Classifier")
plt.xlabel("Predikterad")
plt.ylabel("Sant")
plt.show()
print(classification_report(y_test_filtered, gradient_boosting_classifier_pred))
````

### Random Forest Classifier

````python
random_forest_classifier = RandomForestClassifier(random_state=42, n_jobs=-1)
random_forest_classifier.fit(X_train_scaled_filtered, y_train_filtered)
random_forest_classifier_pred = random_forest_classifier.predict(X_test_scaled_filtered)
````

````python
cm_gb = confusion_matrix(y_test_filtered, random_forest_classifier_pred)
plt.figure(figsize=(6,5))
sns.heatmap(cm_gb, annot=True, fmt="d", cmap="Reds")
plt.title("Confusion Matrix - Random Forest Classifier")
plt.xlabel("Predikterad")
plt.ylabel("Sant")
plt.show()
print(classification_report(y_test_filtered, random_forest_classifier_pred))
````

### Extra Trees Classifier

````python
extra_trees_classifier= ExtraTreesClassifier(random_state=42, n_jobs=-1)
extra_trees_classifier.fit(X_train_scaled_filtered, y_train_filtered)
extra_trees_classifier_pred = extra_trees_classifier.predict(X_test_scaled_filtered)
````

````python
cm_gb = confusion_matrix(y_test_filtered, extra_trees_classifier_pred)
plt.figure(figsize=(6,5))
sns.heatmap(cm_gb, annot=True, fmt="d", cmap="Greens")
plt.title("Confusion Matrix - Extra Trees Classifier")
plt.xlabel("Predikterad")
plt.ylabel("Sant")
plt.show()
print(classification_report(y_test_filtered, extra_trees_classifier_pred))
````

### Utvärdering av modeller i default-utförande

Gradient Boosting Classifier presterar bäst på alla punkter, följd av Random Forest Classifier och sist Extra Trees Classifier.

## Hyper Parameter Tuning

För att utröna om där finns mer optimala parametrar att träna modellen med genomför jag en hyper parameter tuning (HPT). Genom att använda GridSearchCV med flera olika parametrar och sedan justera efter resultatet och köra igen kan jag finjustera vilka parametrar jag ska skicka med i modellerna för att eventuellt kunna förbättra resultatet. Nedan kan vi se de olika inställningarna jag valt att testa. 
- n_estimators är antalet träd i en random forest modell. Att hitta ett optimalt antal träd kan förbättra modellens prestanda.
- learning_rate bestämmer hur snabbt modellen lär sig (endast applicerbart på GradientBoostingClassifier).
- max_depth anger det maximala antalet nivåer i ett beslutsträd.
